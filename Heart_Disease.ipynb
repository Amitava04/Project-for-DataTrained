```python
df.skew()
```

Some data are skewed and it's we can not use logarithm fuction to remove skewness on our target column

So we make a new dataframe to use logarithm function

```python
new=df.copy()
```

```python
new=df.drop(columns=['Target'])
```

```python
new.columns
```

```python
for col in new.columns:
    if new[col].skew()>abs(0.55):
        new[col]=np.log1p(new[col])
```

```python
new.skew()
```

```python
x=new
```

```python
y=df['Target']
y.value_counts()
```

```python
columns_Scale=['Age','trestbps','chol','thalach','oldpeak']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x[columns_Scale]=scaler.fit_transform(x[columns_Scale])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.80,random_state=49)
```

OBSERVATION: As our dataset has very small amout of data for a multinomial problem so we take only 20% for testing and 80% for the training

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
model=[DecisionTreeClassifier(),RandomForestClassifier(),SVC(),KNeighborsClassifier()]
```

```python
for m in model:
    m.fit(x_train,y_train)
    print('Score of',m,'is',m.score(x_train,y_train))
    pred=m.predict(x_test)
    print('Confusion Matrix',confusion_matrix(pred,y_test))
    print('Accuracy Score',accuracy_score(pred,y_test))
    print('Classification Report'),classification_report(pred,y_test)
    print('******************************************************************************************')
    print('\n')
```


Saving the best model with accuracy 32.5% from RandomForest. It's a multinomial class problem and with these small number of data it can't show more accuracy than this. The random probability of predicting one out of our 5 values is 20% In that scenario our model is trained well to show 20% better accuracy than randomness with this small amount of data. So the accuracy score is not that bad

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(RandomForestClassifier(),x,y,cv=4,scoring='r2')
print('Score is',score)
print('Mean score',score.mean())
print('Standard Deviation',score.std())
```

```python
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
pred=rfc.predict(x_test)
plt.scatter(y_test,pred)
plt.xlabel('x_test')
plt.ylabel('predicted Y')
plt.title('Prediction of Heart Disease')
plt.show()
```
