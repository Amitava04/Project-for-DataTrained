```python
train=pd.get_dummies(train,columns=['TITLE','CUISINES','CITY','CITY','LOCALITY'],prefix=['TITLE','CUISINES','CITY','CITY','LOCALITY'])
```

```python
train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
y=train['COST']
```

```python
x=train.drop('COST',axis=1)
```

```python
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=120)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
model=[RandomForestRegressor()]
```

```python
for m in model:
    m.fit(x_train,y_train)
    print('The score of',m,'is',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print('Errors')
    print('Mean Squared Error',mean_squared_error(y_test,predm))
    print('Mean Absolute Error',mean_absolute_error(y_test,predm))
    print('R2 Score',r2_score(y_test,predm))
    print('*************************************************************************')
    print('\n')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(RandomForestRegressor(),x,y,cv=7,scoring='r2')
print('Score is',score)
print('Mean Score',score.mean)
print('Standard Deviation',score.std)
```
